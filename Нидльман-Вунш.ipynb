```python
def align_nuc(s:str, g:str, gap_penalty=1):   
        F = [[0] * (len(g) + 1) for i in range(len(s) + 1)]
        mismatch_penalty=gap_penalty*2
        for i in range(len(s)+1):
            F[i][0] = i*gap_penalty
        for i in range(len(g) + 1):
            F[0][i] = i*gap_penalty
        s1 = ' ' + s
        g1 = ' ' + g
        for i in range(1, len(s) + 1):
            for j in range(1, len(g) + 1):
                if s1[i] == g1[j]:
                    F[i][j] = min(F[i - 1][j - 1]-mismatch_penalty, F[i-1][j]+gap_penalty, F[i][j-1]+gap_penalty)
                else:
                    F[i][j] = min(F[i][j - 1]+gap_penalty, F[i -1][j-1]+mismatch_penalty, F[i-1][j]+gap_penalty)
        i=len(s1)-1
        j=len(g1)-1
        s_res=""
        g_res=""
        while i > 0 and j > 0:
            if F[i][j]==F[i-1][j-1]+mismatch_penalty and s1[i]!=g1[j]:
                s_res+=s1[i]
                g_res+=g1[j]
                i-=1
                j-=1
            if F[i][j]==F[i-1][j-1]-mismatch_penalty and s1[i]==g1[j]:
                s_res+=s1[i]
                g_res+=g1[j]
                i-=1
                j-=1
            if F[i][j]==F[i-1][j]+gap_penalty:
                s_res+=s1[i]
                g_res+="_"
                i-=1
            if F[i][j]==F[i][j-1]+gap_penalty:
                s_res+="_"
                g_res+=g1[j]
                j-=1
        if i==0 and j!=0:
            for x in range(j, 0, -1):
                g_res+=g1[x]
                s_res+="_"
        if i!=0 and j==0:
            for x in range(j, 0, -1):
                s_res+=s1[x]
                g_res+="_"
        s_res=s_res[::-1]
        g_res=g_res[::-1]
        print(s_res)
        print(g_res)
    
    
str1 = input()
str2 = input()
gap_penalty=int(input())
align_nuc(str1, str2, gap_penalty)
```
